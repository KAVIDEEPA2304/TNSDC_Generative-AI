```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define data for autumn and winter temperatures
months = np.arange(1, 13)
autumn_temperatures = np.array([20, 18, 15, 12, 10, 8, 10, 12, 15, 18, 20])
winter_temperatures = np.array([10, 8, 5, 2, 0, -2, 0, 2, 5, 8, 10])

# Create a meshgrid for months and temperatures
months_grid, temp_grid = np.meshgrid(months, np.concatenate((autumn_temperatures, winter_temperatures)))

# Create a meshgrid for seasons (0 for autumn, 1 for winter)
seasons_grid = np.zeros_like(months_grid)
seasons_grid[:, len(autumn_temperatures):] = 1

# Plot the 3D surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot autumn temperatures
ax.plot_surface(months_grid, temp_grid, seasons_grid, color='orange', alpha=0.7)

# Plot winter temperatures
ax.plot_surface(months_grid, temp_grid, seasons_grid + 1, color='blue', alpha=0.7)

# Set labels and title
ax.set_xlabel('Months')
ax.set_ylabel('Temperature (°C)')
ax.set_zlabel('Season')
ax.set_title('Transition from Autumn to Winter')

# Set z ticks
ax.set_zticks([0, 1])
ax.set_zticklabels(['Autumn', 'Winter'])

# Show the plot
plt.show()

```
